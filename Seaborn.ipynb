## Hist plot :

## Scatter Plot

```python
sns.scatterplot(x = iris['petal_length'], y = iris['petal_width']) # reletionship bt two variable
plt.show()
```

```python
# sns.scatterplot(x = iris['petal_length'], y = iris['petal_width']) 
# or
sns.scatterplot(x = 'petal_length', y = 'petal_width', data=iris, hue = 'species') # hue -> for color coding

plt.show()
```

## Joint Plot

```python
sns.jointplot(data=iris, x = 'petal_length', y = 'petal_width') # histogram/ distribution and reletion bt variable together
plt.show()
```

```python
sns.jointplot(data=iris, x = 'petal_length', y = 'petal_width', hue='species')
plt.show()
```

```python
sns.jointplot(data=iris, x = 'petal_length', y = 'petal_width', kind='reg') #Use of `hue` with `kind='reg'` is not currently supported
plt.show()
```

```python
sns.jointplot(data=iris, x = 'petal_length', y = 'petal_width', kind='kde') 
plt.show()
```

```python
sns.jointplot(data=iris, x = 'petal_length', y = 'petal_width', kind='kde', hue='species') 
plt.show()
```

## Choosing the right visualisation/plot

Q1. What is the right chart to use to find dependency between 
one Continuous Variable and Categorical variable?

ans. count plot/ boxplot

------------------
**1D viz** : one variable -- num distribution (hist, kde,boxplot), categorical, count (countplot, countplot) 

**2D viz** : two variables--num-num (scatterplot, jointplot), cat-cat (??), cat-num (countplot, boxplot) 

**2D viz** : multiple variables -- pairplot, heatmap. -

**Multi-dimensional viz** : scatter plot with hue(color coding), num-num-categorical,cat-cat-num(barplot with hue) surface plot

```python
# 1D plots
sns.histplot(iris['sepal_length'], bins = 20)
```

```python
setosa = iris.loc[iris['species']=='setosa']
sns.boxplot(data = setosa, y = 'sepal_length')
```

```python
setosa = iris.loc[iris['species']=='setosa']
sns.boxplot(data = iris, y = 'sepal_length')
```
