## Subplot

````python
x = np.arange(1,10,0.1)
y = np.sin(x)

plt.plot(x,y)
````

````python
x = np.arange(1,10,0.1)
y = np.sin(x)

plt.scatter(x,y)
````

````python
x = np.arange(1,10,0.1)
y = np.sin(x)

plt.bar(x,y)
````

````python
# Ploting sin(x) and tan(x)  together
x = np.arange(1,10,0.1)
y1 = np.sin(x)
y2 = np.tan(x)

fig, ax = plt.subplots(2, 1)
ax[0].plot(x, y1)
ax[1].plot(x, y2)

plt.show()
````

````python
# Ploting sin(x), tan(x) and log(x)  together
x = np.arange(1,10,0.1)
y1 = np.sin(x)
y2 = np.tan(x)
y3 = np.log(x)

fig, ax = plt.subplots(3, 1)
ax[0].plot(x, y1)
ax[1].plot(x, y2)
ax[2].plot(x, y3)

plt.show()
````

````python
# Ploting sin(x), tan(x) and log(x)  together
x = np.arange(1,10,0.1)
y1 = np.sin(x)
y2 = np.tan(x)
y3 = np.log(x)

fig, ax = plt.subplots(1, 3) # ploting in 1 row and 3 column
ax[0].plot(x, y1)
ax[1].plot(x, y2)
ax[2].plot(x, y3)

plt.show()
````

````python
# Ploting sin(x), tan(x) and log(x)  together
x = np.arange(1,10,0.1)
y1 = np.sin(x)
y2 = np.tan(x)
y3 = np.log(x)

plt.figure()

plt.subplot(2,3,1) # remember that indexing starts from 1 here
plt.plot(x,y1)

plt.subplot(2,3,3)
plt.plot(x,y2)

plt.subplot(2,3,5)
plt.plot(x,y3)

plt.show()
````

````python
# Another method for subplot

fig = plt.figure()
ax231 = fig.add_subplot(231)
ax231.plot(x, y1)
ax233 = fig.add_subplot(233)
ax233.plot(x, y2)
ax235 = fig.add_subplot(235)
ax235.plot(x, y3)
plt.show()
````

## 3-D Plot

````python
a = np.array([0,1,2,3])
b = np.array([0,1,2])
a,b = np.meshgrid(a,b)
````

````python
a
````

````python
b
````

````python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(a, b, np.array([[7]])) # 3rd argument is the elevation in 3-D we want
plt.show() 
````

````python
a = np.arange(-1, 1, 0.005)
b=a
a,b = np.meshgrid(a,b)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(a, b, a**2 + b**2)
plt.show()
````
